```python
from kiwipiepy import Kiwi

kiwi = Kiwi()
def find_nouns_verb_adject(text):
    """입력된 문장에서 명사, 동사 추출해 리스트로 반환"""
    tokens = kiwi.tokenize(text)

    target_idxs = []
    target_words = []
    for idx, wrd in enumerate(tokens):
        if 'NN' in wrd.tag:
            if len(wrd.form) > 1:
                target_words.append(wrd.form)
                target_idxs.append('noun')
        elif wrd.tag.startswith("V"):
            tmp_form = wrd.form + "다" 
            target_words.append(tmp_form)
            target_idxs.append('verb')
    return target_words, target_idxs

def find_intension_one_text(words, words_tag):
    """문장 내 의도를 나타내는 단어를 추출한다"""
    intensions = []
    for idx, tag in enumerate(words_tag):
        try:
            if tag == 'verb':
                word = "".join(word[idx-1:idx+1])
                intensions.append(word)
            elif(tag=='noun') and (words_tag[idx+1]!='verb'):
                intensions.append(words[idx])
        except:
            intensions.append(words[idx])
    return intensions

def find_intension(text):
    """입력된 문장에서 의도를 찾아서 반환한다"""
    target_words, target_idxs = find_nouns_verb_adject(text)
    intensions = find_intension_one_text(target_words, target_idxs)
    return intensions
```

```python
#자주 등장하는 NNG,VV,NNP,VA만 분석
def find_nouns(keyword):
    """명사 태그 반환"""
    tokens = kiwi.tokenize(keyword)
    for token in tokens:
        if 'NN' in token.tag:
            return 'Noun'
        elif 'VV' in token.tag:
            return 'VV'
        elif 'VA' in token.tag:
            return 'VA'
        else:
            break
    return 'ETC'
df_key_count['tag'] = df_key_count['keyword'].apply(lambda x:find_nouns(x))
df_key_count
```

```python
from collections import Counter

# 모든 문장을 하나로 결합
df_onesent = " ".join(df['only_kor_contents'].tolist())

# 품사 분석
pos_result = komoran.pos(df_onesent)

# 명사(NNP, NNG), 동사(VV), 형용사(VA)만 추출 (길이 > 1인 것만)
allowed_pos = ['NNP', 'NNG', 'VV', 'VA']
words = [word for word, tag in pos_result if tag in allowed_pos and len(word) >= 1]

# 가장 많이 등장한 상위 10개
Counter(words).most_common(10)
```

```python
remove_words = {'영어', '초대', '화상', '수업', '원어민', '후기', '회화', '대화', '무료', '프로모션', '할인', '코드', '선물', '친구'}
nouns_df = [i for i in words if i not in remove_words]
```

```python
nouns_df
```

```python
def getGram(tokens, n):
    """분절된 단어를 전달하면 nGram을 반환"""
    bigrams = []

    for i in range(0, int(len(tokens)/n)):
        bigrams.append(tokens[i:i+n])

    ngram_result = pd.DataFrame(data={'nGram':bigrams})
    ngram_result['nGram'] = ngram_result['nGram'].apply(lambda x:" ".join(x))
    ngram_result = ngram_result['nGram'].value_counts().to_frame()
    ngram_result.reset_index(inplace=True)

    return ngram_result
n2gram_result = getGram(tokens=nouns_df, n=2)
n3gram_result = getGram(tokens=nouns_df, n=3)

display(n2gram_result[:15])
display(n3gram_result[:15])
```

```python
import torch
from transformers import pipeline
import pandas as pd

# 엑셀 파일 읽기
df = pd.read_excel("review_data.xlsx")

# 텍스트 연결
full_text = " ".join(df['only_kor_contents'].dropna().tolist())

# 요약기 로드
summarizer = pipeline("summarization", model="sshleifer/distilbart-cnn-12-6")

# 텍스트를 1000자 단위로 분할
chunks = [full_text[i:i+1000] for i in range(0, len(full_text), 1000)]

# 각 chunk 요약
summaries = [summarizer(chunk, max_length=80, min_length=30, do_sample=False)[0]['summary_text'] for chunk in chunks]

# 최종 요약
final_summary = " ".join(summaries)
print(final_summary)
```
